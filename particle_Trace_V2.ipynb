```python
import matplotlib.pyplot as plt
import plotly
import plotly.graph_objs as go
import plotly.express as px
import numpy as np
import random
```

```python
def sig ():
    return random.choice([-1, 1])

def paso_3D (N: int, item: int):
    direcciones = np.array([[1, 0, 0], [0, sig(), 0], [0, 0, sig()], [sig(), sig(), sig()], [sig(), sig(), 0], [sig(), 0, sig()], [0, sig(), sig()]])
    if item < N/2:
        proba = (N-item)/N
        probb = (1- proba)/6
        si = direcciones[np.random.choice([0,1,2,3,4,5,6], p = [proba, probb, probb, probb, probb, probb, probb])]
    elif item >= N/2:
        proba = 0.25
        probb = 0.75/6
        si = direcciones[np.random.choice([0,1,2,3,4,5,6], p = [proba, probb, probb, probb, probb, probb, probb])]
    return si
```

```python
def walker_3D (N):
    xs = np.zeros(N, dtype = np.int64)
    ys = np.zeros(N, dtype = np.int64)
    zs = np.zeros(N, dtype = np.int64)
    place = np.array([0,0,0])
    for i in range (1, N):
        place = place + paso_3D(N, i)
        xs[i] = place[0]
        ys[i] = place[1]
        zs[i] = place[2]
    return xs, ys, zs
```

```python
N = 500
xs, ys, zs = walker_3D(N)
tiempo = np.array(range(0,N))
fig = go.Figure(data = go.Scatter3d(
    x = xs, y = ys, z = zs,
    marker = dict(
        size = 4,
        color = tiempo,
        colorscale = px.colors.sequential.Plasma,
    ),
    line = dict(
        color = "darkblue",
        width = 2
    )
))
fig.update_layout(scene=dict(
    yaxis = dict(range = [-50, 50]),
    zaxis = dict(range = [-50, 50])
))
fig.show()
```
