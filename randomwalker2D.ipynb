```python
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
import numpy as np
import random
```

```python
def sig ():
    a = random.choice([-1, 1])
    return a

def paso_2D():
    direcciones = np.array([[sig(), 0], [0, sig()]])
    return direcciones[np.random.choice([0, 1])]
```

```python
def walker_2D (N):
    xs = np.zeros(N, dtype = np.int64)
    ys = np.zeros(N, dtype = np.int64)
    place = np.array([0,0])
    resul = 0
    for i in range(1,N):
        place = place + paso_2D()
        xs[i] = place[0]
        ys[i] = place[1]
        if (xs[i] == 0) and (ys[i] == 0):
            resul = resul + 1
        else:
            None
    return xs, ys, resul
```

```python
N = 500000
xs, ys, resul = walker_2D(N)
fig, ax = plt.subplots(figsize = (10, 10))
ax.plot(xs, ys)
print("The random walk goues through the origin: " + str(resul) + " times")
```

```python
N = 500000
n = 50
color_idx = np.linspace(1, 0, n)
fig, ax = plt.subplots(figsize=(15,15))
cant = 0
for i in range (0, n):
    xs, ys, resul = walker_2D(N)
    ax.plot(xs, ys, color = plt.cm.gist_rainbow(color_idx[i]))
    if (resul != 0):
        cant = cant + 1
    else:
        None
por = (cant/n) * 100
print("El porcentaje de random walks que pasa por el origen nuevamente es de " + str(por) + "%")
```
